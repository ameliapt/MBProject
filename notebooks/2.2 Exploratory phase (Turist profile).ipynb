#### Statistic: Número de turistas según comunidad autónoma de destino principal

```python
file_name = 'num_turists_by_ccaa.xlsx'
path = 'C:/Users/ameli/IRONHACK/WEEK_5/MBP/data/clean/'

num_turists_by_ccaa = pd.read_excel(path+file_name)
num_turists_by_ccaa = num_turists_by_ccaa.drop('Unnamed: 0', axis=1)
num_turists_by_ccaa.head()
```

```python
num_turists_by_ccaa_summary = describe_df(num_turists_by_ccaa)
```

```python
num_turists_by_ccaa_hist = create_histograms(num_turists_by_ccaa,'destiny', 'value')
```

```python
num_turists_by_ccaa_boxplots = create_boxplots(num_turists_by_ccaa,'destiny', 'value')
```

```python
results_num_tur = pd.crosstab(index= num_turists_by_ccaa.destiny, columns = num_turists_by_ccaa.date, values = num_turists_by_ccaa.value, aggfunc = 'mean')
display(round(results_num_tur))
```

#### Statistic: Número de turistas según tipo de alojamiento principal

```python
file_name = 'num_turists_by_stay.xlsx'
path = 'C:/Users/ameli/IRONHACK/WEEK_5/MBP/data/clean/'

num_tur_stay_full = pd.read_excel(path+file_name)
num_tur_stay_full = num_tur_stay_full.drop('Unnamed: 0', axis=1)
num_tur_stay_full.head()
```
