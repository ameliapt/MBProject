# Spain's tourism industry dashboard

### Exploratory Data Analysis I

##### What is the data telling us?

The main goal of this Exploratory Data Analysis is to check if the data that has been retrieved from the Instituto Nacional de Estadística (INE) and cleaned afterward can be directly used to create visualizations or, on the contrary, further processing is needed. 

The EDA process has been divided into three notebooks:

- 2.1 Tourist Expenditure
- 2.2 Tourist Profile
- 2.3 Hotel and apartments market

All three notebooks follow the same procedure: to explore the data, I have defined a function that summarizes the main characteristics of each data frame. Depending on the information stored on them, I have created some histograms and boxplots to check the existence of outliers.

Regarding outliers: the data from INE has been collected and classified by dates. It means that each date contains information such as numeric values and categorical labels. For example, one data frame can contain information about one specific country and, at five rows below, store information about total expenditure in Spain. 

For this reason, most boxplots show outliers that I have manually checked and concluded are not outliers. The distribution of the numerical data can be observed pleasantly with the tables created below. 

Data has not been scaled or transformed given that the goal of this project is to create visualizations, not models. You can find the result visualizations and the dashboards created with these data here: https://public.tableau.com/app/profile/amelia3703/viz/MBP

If you want to know more about this project, you can also take a look at the README file in my Github Repository: https://github.com/ameliapt/MBProject
