```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import openpyxl
```

 ##  Turism expenditure in Spain

#### Statistic: Gasto de los turistas internacionales según comunidad autónoma de destino principal y país de residencia


```python
path = 'C:/Users/ameli/IRONHACK/WEEK_5/MBP/data/clean/'
file_name = 'turism_expenditure_by_ccaa_origin.xlsx'

tur_exp_by_ccaa_origin = pd.read_excel(path+file_name, engine = 'openpyxl')
tur_exp_by_ccaa_origin = tur_exp_by_ccaa_origin.drop('Unnamed: 0', axis=1)
tur_exp_by_ccaa_origin.head()
```

```python
def describe_df(df):
    print('Shape of dataframe is:')
    print(df.shape)
    print('')
    print('Data types of dataframe are:') 
    print(df.dtypes)
    print('')
    print('Summary of numeric data is:')
    print(df.describe().T)
    print('')
    print('Summary of categorical data is:')
    print(df.describe(include='object').T)
    print('')
    print('Does it contains missing values?')
    print(df.isna().sum())
    print('')
    print('Value counts:')
    for col in df:
        print(df[col].value_counts(),'\n')
```

```python
summary = describe_df(tur_exp_by_ccaa_origin)
```

### Let's take a look at the data

```python
fig, ax = plt.subplots(dpi=100)
ax = sns.boxplot(tur_exp_by_ccaa_origin.value, palette= 'colorblind')
ax.set_title('Boxplot of turists expenditure by origin and ccaa')
plt.show()
```

It seems that there are some outliers. Let's check if they are indeed outliers or extreme values.

```python
fig, ax = plt.subplots(dpi=100)
ax = sns.histplot(tur_exp_by_ccaa_origin['value'], palette= 'colorblind')
ax.set_title('Histogram of turists expenditure by origin and ccaa')
plt.show()
```

```python
### df
q1 = tur_exp_by_ccaa_origin['value'].quantile(q= 0.25)
q3 = tur_exp_by_ccaa_origin['value'].quantile(q= 0.75)
iqr = q3 - q1

# Upper and lower limit
lower_limit = q1 - 1.5*iqr
upper_limit = q3 + 1.5*iqr

print('Q1 is:', q1)
print('Q3 is:', q3)
print('IQR is:', iqr)
print('Upper limit is:', upper_limit)
print('Lower limit is:', lower_limit)
```

```python
outliers = tur_exp_by_ccaa_origin['value'][(tur_exp_by_ccaa_origin['value'] >= upper_limit) | (tur_exp_by_ccaa_origin['value'] <= lower_limit)]
print(outliers.sort_values(ascending=False))
```

This is tricky: plots are showing us outliers because numeric data are not in the same unit (mixed days with millions). Let's look closer.

```python
days = tur_exp_by_ccaa_origin['value'].loc[(tur_exp_by_ccaa_origin['category'] == 'Duración media de los viajes') == True]
fig, ax = plt.subplots(dpi=100)
ax = sns.histplot(days, palette= 'colorblind')
ax.set_title('Turists expenditure by days')
plt.show()
```

```python
gmp = tur_exp_by_ccaa_origin['value'].loc[(tur_exp_by_ccaa_origin['category'] == 'Gasto medio por persona') == True]
fig, ax = plt.subplots(dpi=100)
ax = sns.histplot(gmp, palette= 'colorblind')
ax.set_title('Mean expenditure per person')
plt.show()
```

```python
gmdp = tur_exp_by_ccaa_origin['value'].loc[(tur_exp_by_ccaa_origin['category'] == 'Gasto medio diario por persona') == True]
fig, ax = plt.subplots(dpi=100)
ax = sns.histplot(gmp, palette= 'colorblind')
ax.set_title('Daily mean expenditure per person')
plt.show()
```

```python
gt = tur_exp_by_ccaa_origin['value'].loc[(tur_exp_by_ccaa_origin['category'] == 'Gasto total') == True]
fig, ax = plt.subplots(dpi=100)
ax = sns.histplot(gt, palette= 'colorblind')
ax.set_title('Total expenditure by country')
plt.show()
```

In the following table we can see clearly the mean values.

```python
mean_by_origin = pd.crosstab(index= tur_exp_by_ccaa_origin.origin, columns = tur_exp_by_ccaa_origin.category, values = tur_exp_by_ccaa_origin.value, aggfunc = 'mean')
display(round(mean_by_origin))
```

```python
gby_mean = tur_exp_by_ccaa_origin.groupby(['origin', 'category','date']).mean()
gby_mean
```

#### Statistic: Gasto de los turistas internacionales según partidas de gasto y comunidad autónoma de destino principal


```python
path = 'C:/Users/ameli/IRONHACK/WEEK_5/MBP/data/clean/'
file_name = 'turism_expenditure_by_ccaa_item.xlsx'

tur_exp_by_ccaa_item = pd.read_excel(path+file_name, engine = 'openpyxl')
tur_exp_by_ccaa_item = tur_exp_by_ccaa_item.drop('Unnamed: 0', axis=1)
tur_exp_by_ccaa_item.head()
```

```python
summary = describe_df(tur_exp_by_ccaa_item)
```

```python
def create_histograms(df, col_name_category, col_name_values):
    fig, ax = plt.subplots(dpi=100)
    for i in df[col_name_category].unique():
        item = df[col_name_values].loc[(df[col_name_category] == i) == True]
        ax = sns.histplot(item, palette= 'colorblind')
        ax.set_title('Histogram of {}'.format(i))
        plt.show()
        
def create_boxplots(df, col_name_category, col_name_values):
    fig, ax = plt.subplots(dpi=100)
    for i in df[col_name_category].unique():
        item = df[col_name_values].loc[(df[col_name_category] == i) == True]
        ax = sns.boxplot(item, palette= 'colorblind')
        ax.set_title('Boxplot of {}'.format(i))
        plt.show()
```

```python
histograms_tur_exp_by_ccaa_item = create_histograms(tur_exp_by_ccaa_item, 'expense_item', 'value')
```

```python
boxplots_tur_exp_by_ccaa_item = create_boxplots(tur_exp_by_ccaa_item, 'expense_item', 'value')
```
